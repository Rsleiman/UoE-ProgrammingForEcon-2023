## Exercise 4:

```python
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)

X, Y = np.meshgrid(x, y)

Z = np.log(X) + np.sin(Y)

# Creating a axes object
ax = plt.axes(projection="3d")

# Inputting plot data into axes object
ax.contour3D(X, Y, Z, 100)

# Assigning labels
ax.set_xlabel("x", fontweight="heavy")
ax.set_ylabel("y", fontweight="heavy")
ax.set_zlabel("z", fontweight="heavy")
ax.set_title("z = log(x) + sin(y)", fontweight="heavy")

plt.show()
```

## Exercise 5:

```python
def objective_function(x, alpha):
    return -(x[0] ** alpha * x[1] ** (1 - alpha))


def constraint(x, p1, p2, y):
    return y - (p1 * x[0] + p2 * x[1])


# Parameter values
alpha = 0.5
y = 10
p1 = 1
p2 = 2

# defining 'minimize' arguments:
# x0:
xinit = [randint(1, 10), randint(1, 10)]
# bounds:
bnds = [(0, None), (0, None)]
# constraints:
cons = [{"type": "ineq", "fun": constraint, "args": (p1, p2, y)}]


res = scipy.optimize.minimize(
    objective_function, x0=xinit, args=alpha, bounds=bnds, constraints=cons
)

print(f"The optimum amount of good x1 is {round(res.x[0], 3)}")
print(f"The optimum amount of good x2 is {round(res.x[1], 3)}")
print(f"The resulting maximum utility is {round(-res.fun,3)}")
```
