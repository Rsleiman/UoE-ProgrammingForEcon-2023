## Exercise 5

Part a:

```python
def Y(k, l, a, alpha, sigma):
    # Error checking
    if a > 0 and 0 < alpha < 1 and sigma > 0:
        if sigma == 1:      # Cobb-Douglas
            return a*(k**alpha)*(l**(1-alpha))
        else:               # CES
            return a*(alpha*(k**((sigma-1)/sigma))+(1-alpha)*(l**((sigma-1)/sigma)))**(sigma/(sigma-1))
    else:
        raise ValueError("One or more arguments are invalid")

a = 1.5
alpha1 = 0.33
```

Part b:

```python
sigma1 = 1
l = 3

k_grid = np.linspace(0,10,11)    # Assign x grid
# y_grid = np.zeros(11)            # Assign y grid
# for i in range(y_grid.size):     # Change y values to Y
#     y[i] = Y(k_grid[i], l, a, alpha, sigma)

Y1 = Y(k_grid, l, a, alpha1, sigma1)

fig, ax = plt.subplots()    # Create figure and axis objects
ax.plot(k_grid,Y1)                # Plot values to axis
    # Set labels
ax.set_xlabel("Capital, K")    
ax.set_ylabel("Production, Y")
ax.set_title("CES Production Function, Ïƒ = 1")

plt.show()                  # Display graph
```

Part f:

Sigma represents the elasticity of substitution, which is a measure of how easily capital can be substituted for labour in the production process. 

A higher elasticity of substitution means that it is easier to substitute capital for labour, while a lower elasticity of substitution makes this more difficult.

More specifically, for a fixed level of capital and alpha, a higher sigma would indicate that labour exhibits lower diminishing marginal returns.

This can be seen on the graph from part e, where a higher sigma results in higher production at every level of fixed capital because more output can be produced using capital for a given level of labour.

With a lower level of sigma, diminishing marginal returns has a more significant effect. We can also see this on the graph: at capital levels further from labour = 3, the differences in production between the different sigma levels increase.
